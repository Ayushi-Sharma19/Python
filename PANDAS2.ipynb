```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing

data = fetch_california_housing()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['Target'] = data.target

print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())
print("\nData info:")
print(df.info())
print("\nStatistical summary:")
print(df.describe())

print("\nFirst 5 rows, last 3 columns:")
print(df.iloc[:5, -3:])
print("\nRows 50-60 for specific columns:")
print(df.loc[50:60, ['MedInc', 'AveRooms', 'Target']])

high_income = df[df['MedInc'] > 5]
low_rooms = df[df['AveRooms'] < 5]

df['Half_MedInc'] = df['MedInc'] * 0.5
df['Price_per_Room'] = df['Target'] / df['AveRooms']

df.drop(['AveOccup'], axis=1, inplace=True)

print("\nTotal Median Income:", df['MedInc'].sum())
print("Lowest Target Value:", df['Target'].min())
print("Highest Target Value:", df['Target'].max())

df['Income_Category'] = df['MedInc'].apply(lambda x: 'High' if x > 5 else 'Low')
df['Expensive'] = df.apply(lambda row: 'Yes' if row['Target'] > 3 else 'No', axis=1)

print("\nModified dataframe:")
print(df.head())
```
